<h3 dir='rtl'>توزیع اطلاعات شرکت کنندگان(سن، وزن، قد، جنسیت و میزان مصرف اینترنت) </h3>

```python
# columns = ['Age', 'Physical-Weight', 'Physical-Height', 'Sex', 'PreInt_EduHx-computerinternet_hoursday']
# df_cleaned = hbn_data[columns].dropna()

fig, axes = plt.subplots(2, 3, figsize=(16, 9), constrained_layout=True)

sns.histplot(hbn_data['Age'], kde=True, ax=axes[0, 0], color='navy')
axes[0, 0].set_title("Age Distribution")
axes[0, 0].set_xlabel("Age")
axes[0, 0].set_ylabel("Count")


sns.histplot(hbn_data['Physical-Weight'], kde=True, ax=axes[0, 1], color='darkcyan')
axes[0, 1].set_title("Weight Distribution")
axes[0, 1].set_xlabel("Weight (lbs)")
axes[0, 1].set_ylabel("Count")

sns.histplot(hbn_data['Physical-Height'], kde=True, ax=axes[0, 2], color='salmon')
axes[0, 2].set_title("Height Distribution")
axes[0, 2].set_xlabel("Height (inches)")
axes[0, 2].set_ylabel("Count")

sns.countplot(x='Sex', data=hbn_data, ax=axes[1, 0], palette="pastel", hue='Sex', dodge=False, legend=False)
axes[1, 0].set_title("Sex Distribution")
axes[1, 0].set_xlabel("Sex")
axes[1, 0].set_ylabel("Count")

sns.histplot(hbn_data['PreInt_EduHx-computerinternet_hoursday'], kde=True, ax=axes[1, 1], color='purple')
axes[1, 1].set_title("Internet Usage Distribution")
axes[1, 1].set_xlabel("Hours per Day")
axes[1, 1].set_ylabel("Count")


axes[1, 2].axis('off')



plt.show()

```

<h3 dir='rtl'>توزیع توده‌ی عضلانی، نرخ متابولیسم پایه و میزان عملکرد ذهنی را بر حسب جنسیت</h3>

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5)) 

 
sns.kdeplot(data=hbn_data, x="BIA-BIA_BMI", hue="Sex", fill=True, ax=axes[0], palette='muted')
axes[0].set_title('Distribution of BMI by Gender', fontsize=14)
axes[0].set_xlabel('BMI', fontsize=12)
axes[0].set_ylabel('Density', fontsize=12)


sns.kdeplot(data=hbn_data, x='BIA-BIA_BMR', hue='Sex', fill=True, ax=axes[1], palette='muted')
axes[1].set_title('Distribution of BMR by Gender', fontsize=14)
axes[1].set_xlabel('BMR', fontsize=12)
axes[1].set_ylabel('Density', fontsize=12)

sns.kdeplot(data=hbn_data, x='CGAS-Score', hue='Sex', fill=True, ax=axes[2], palette='muted')
axes[2].set_title('Distribution of CGAS Score by Gender', fontsize=14)
axes[2].set_xlabel('CGAS Score', fontsize=12)
axes[2].set_ylabel('Density', fontsize=12)






plt.show()
```

```python
variables = ['BIA-BIA_BMI', 'BIA-BIA_BMR', 'CGAS-Score']

for var in variables:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Sex', y=var, data=hbn_data, palette='Set2')
    plt.title(f'{var.replace("_", " ").capitalize()} Distribution by Gender')
    plt.xlabel('Gender')
    plt.ylabel(var.replace("_", " ").capitalize())
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

plt.show()
```

<h3 dir='rtl'>توزیع میانگین میزان تناسب در تست آمادگی جسمانی بر حسب سن</h3>

```python
fitness_columns = [
    'FGC-FGC_PU_Zone', 'FGC-FGC_CU_Zone', 'FGC-FGC_GSD_Zone',
    'FGC-FGC_GSND_Zone', 'FGC-FGC_SRL_Zone', 'FGC-FGC_SRR_Zone', 'FGC-FGC_TL_Zone'
]

hbn_data['Fitness_Mean'] = hbn_data[fitness_columns].mean(axis=1)
mean_fitness_by_age = hbn_data.groupby('Age')['Fitness_Mean'].mean().reset_index()

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

sns.scatterplot(data=hbn_data, x='Age', y='Fitness_Mean', color='blue', ax=axes[0], alpha=0.6)
axes[0].set_title('Distribution of fitness mean by age', fontsize=14)
axes[0].set_xlabel('Age', fontsize=12)
axes[0].set_ylabel('Fitness_Mean', fontsize=12)

sns.lineplot(data=mean_fitness_by_age, x='Age', y='Fitness_Mean', marker='o',  ax=axes[1], color='blue')
axes[1].set_title('Distribution of fitness mean by age', fontsize=14)
axes[1].set_xlabel('Age', fontsize=12)
axes[1].set_ylabel('Fitness_Mean', fontsize=12)

plt.show()


```

<h3 dir='rtl'>توزیع میانگین نور محیطی در طول روز را برای دو گروه دارای اختلال خواب و بدون اختلال خواب</h3>

```python
series_data = pd.read_parquet('series.parquet')
df = pd.merge(hbn_data[['id', 'SDS-SDS_Total_T']], series_data[['id', 'light', 'time_of_day']], on='id')

df['sleep_disorder'] = df['SDS-SDS_Total_T'].apply(lambda x: 'No Sleep Disorder' if x < 50 else 'Has Sleep Disorder')
avg_light = df.groupby(['sleep_disorder', 'time_of_day'])['light'].mean().reset_index()

plt.figure(figsize=(10, 6))
for label in avg_light['sleep_disorder'].unique():
    subset = avg_light[avg_light['sleep_disorder'] == label]
    plt.plot(subset['time_of_day'], subset['light'], label=label)


plt.title('Average Environmental Light Distribution by Sleep Disorder Category')
plt.xlabel('Time of Day')
plt.ylabel('Average Light Level')
plt.legend()
plt.grid(True)
plt.show()


```

```python
light_data = df.groupby('sleep_disorder')['light'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x='sleep_disorder', y='light', data=light_data)
plt.title('Average Daytime Light by Sleep Disorder Status')
plt.xlabel('Sleep Disorder Status')
plt.ylabel('Average Light')
plt.show()
```

<h3 dir='rtl'> ماتریس هم‌بستگی را برای سن، شاخص توده‌ی بدنی، تست قدرت با دست غالب، نرخ متابولیسم پایه، میانگین نور محیطی در روز، مصرف انرژی روزانه، توده‌ی عضله‌ی خالص، میانگین تعداد گام در یک روز و نمره‌ی T اختلال خواب</h3>

```python
df2 = pd.merge(hbn_data, series_data, on='id')
columns = [
    'Age', 'Physical-BMI', 'FGC-FGC_GSD', 'BIA-BIA_BMR', 'light',
    'BIA-BIA_DEE', 'BIA-BIA_SMM', 'step', 'SDS-SDS_Total_T'
]

data = df2[columns]

correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

<h3 dir='rtl'> توزیع تعداد گام‌ها به تفکیک روز هفته</h3>

```python
weekly_steps = series_data.groupby('weekday')['step'].sum()
plt.figure(figsize=(10, 6))
weekly_steps.plot(kind='bar', color='skyblue')
plt.title('Average Steps by weekday')
plt.xlabel('day of week')
plt.ylabel('Average Number of Steps')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='weekday', y='step', data=series_data)
plt.title('Steps Distribution by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Steps')
plt.show()
```

<h3 dir='rtl'> میانگین تعداد گام‌ها به تفکیک فصل</h3>

```python
quarterly_steps = series_data.groupby('quarter')['step'].mean()
plt.figure(figsize=(10, 6))
quarterly_steps.plot(kind='bar', color='salmon')
plt.title('Average Steps by Season')
plt.xlabel('weekday')
plt.ylabel('Average Number of Steps')
plt.show()
```
